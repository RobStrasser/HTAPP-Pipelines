#### Build liger objects

```r
run_groups=c("Lymphoid", "Malignant",   "Myeloid",   "Stromal")
```

```r
tag="_noRP"
#tag=""
```

```r
ribo_genes=fread(file.path(extDir,"ribo_prots.txt"))
rm_genes=c(grep("^MT-",row.names(so@assays$RNA@counts),value = TRUE),ribo_genes$`Approved symbol`)
length(rm_genes)
cat(sort(rm_genes))
```

```r
min_cells=50
run_col="compartments"

for (run_group in run_groups){
    message(run_group)
    if (file.exists(paste0(out,run_group,tag,"_LIGER.RData"))){next}
    sub=subset(so,cells=row.names(so@meta.data[so@meta.data[,run_col]==run_group,])) 
   
    ncells=table(sub@meta.data$name)
    samples=names(ncells[ncells>=min_cells])
    data_list=list()
    for (sample in samples){
        dat=GetAssayData(sub,slot = "counts")[,row.names(sub@meta.data[sub@meta.data$name==sample,])]  
        if (tag=="_noRP"){dat=dat[!row.names(dat)%in%rm_genes,]}
        data_list[[sample]]=dat
    }

    ligerex = createLiger(data_list,take.gene.union = T)
    ligerex = normalize(ligerex)
    ligerex = selectGenes(ligerex,var.thresh = 0.001) # set to lower value than default because within cell type variability might be smaller
    ligerex = scaleNotCenter(ligerex)
    print(length(ligerex@var.genes))

    save(ligerex,file=paste0(out,run_group,tag,"_LIGER.RData"))
    
}
```

#### Suggest k

```r
k_list=list()
for (cell_type in cell_types){
    message(cell_type)
    if (file.exists(paste0(out,cell_type,tag,"_KL.pdf"))){next}
    load(paste0(out,cell_type,tag,"_LIGER.RData"))
    min_cells=min(unlist(lapply(ligerex@raw.data,function(x){dim(x)[2]})))-1
    k_list[[cell_type]]=suggestK(object = ligerex,num.cores = 4, k.test = seq(5, min(50,min_cells), 5))
    pdf(paste0(out,cell_type,tag,"_KL.pdf"),height=4,width=4)
    print(k_list[[cell_type]])
    dev.off()
}
```

#### Run otimizeALS with multiple initialisations 

```r
lda=40 #40#25 #20#10  #5 default according to documentation
k=20 #30
#cell_types
for (run_group in c("Malignant")){
    message(run_group)
    if(file.exists(paste0(out,run_group,tag,"_lambda_",lda,"_k_",k,"_LIGER.RData"))){next}
    load(paste0(out,run_group,tag,"_LIGER.RData"))
    ligerex <- optimizeALS(ligerex, k = k,nrep = 3,lambda = lda)
    defaultW <- getOption("warn")
    options(warn = -1)
    ligerex <- quantile_norm(ligerex)
    options(warn = defaultW)
    save(ligerex, file = paste0(out,run_group,tag,"_lambda_",lda,"_k_",k,"_LIGER.RData")) 
}
```
